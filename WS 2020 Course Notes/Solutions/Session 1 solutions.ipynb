# Solution to Part 1.7 - Guided Plotting of NGRIP oxygen isotope record - *Workshop: 15 minutes*

```
# Load and rename data
dat_NGRIP = pd.read_csv("https://raw.githubusercontent.com/ds4geo/ds4geo/master/data/timeseries/NGRIP_chronology_20.tab", sep="\t", header=20)
dat_NGRIP.columns = ["age", "depth", "error", "d18O"]
```

```
# Plot age vs d18O
plt.plot(dat_NGRIP["age"], dat_NGRIP["d18O"])
plt.xlim(40, 30)
```

# Solution to Part 1.8 - Plotting Alpine flood record from scratch - *Workshop: 20 minutes*

```
# Read and re-name the data, note sep and decimal
dat_flood = pd.read_csv("https://raw.githubusercontent.com/ds4geo/ds4geo/master/data/timeseries/Alps_flood_de.txt", sep=";", header=9, decimal=",")
dat_flood.columns = ["age", "north", "south"]
```

```
# Plot the data
plt.plot(dat_flood["age"], dat_flood["north"], label="north")
plt.plot(dat_flood["age"], dat_flood["south"], label="south")
# Add legend and axes
plt.legend()
plt.xlabel("age (ka)")
plt.ylabel("")
plt.xlim(10,0)
```
